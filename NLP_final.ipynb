## Cleaning The Data

### steps of cleaning the data :
- lowercase the data : to ensure consistency
- Removing URLs
- Handling Contractions , for example change "can't" to "cannot"
- Removing Special Characters 
- Remove punctuation
- Remove Numbers
- stop words : to reduce the input's dimensionality

### Implementation of a function that counts the rows that have  ( URLs , Special Characters , Punctuation , Stop Words , Numbers )

```python
def count_rows(column):
    # Initialize count = zero
    urls_count = 0
    special_chars_count = 0
    punctuation_count = 0
    stop_word_count = 0
    numeric_count = 0
    
    # make a set of all stop words
    stop_words = set(stopwords.words('english'))

    for text in column:
        # Count URLs
        # if URLs found
        if re.search(r'http\S+|www\S+|https\S+', text):
        # add by one
            urls_count += 1

        # Count special characters except english letters and spaces
        # if special characters found 
        if re.search(r'[^A-Za-z0-9\s]+', text):
        # add by one
            special_chars_count += 1
        
        # Count numbers
        # if number found 
        if  re.search(r'\d', text):
        # add by one
            numeric_count +=1

        # Count punctuation
        # if punctuation found 
        if any(char in string.punctuation for char in text):
        # add by one
            punctuation_count += 1

        # Count stop words
        words = text.split()
        # if the word is stop word found 
        if any(word.lower() in stop_words for word in words):
        # add by one
            stop_word_count += 1

    return urls_count, special_chars_count , punctuation_count , stop_word_count , numeric_count
```

### Implementation of a function that lowercase the data , Remove URLs , Handling Contractions , Remove Special Characters , Remove punctuation , Remove Numbers

```python
def clean_data(text):

    # Lowercase the text
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # Handle Contractions
    text = contractions.fix(text)
    
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Remove any special characters except english letters and spaces 
    text=re.sub(r'[^A-Za-z\s]+',' ',text)
    
    # Remove stop words
    # Tokenize the text
    tokens = nltk.word_tokenize(text)
    # create a set of stop words using NLTK libarary
    stop_words = set(nltk.corpus.stopwords.words('english'))
    # iterate over the data and remove any stop word in it
    filtered_tokens = [word for word in tokens if word not in stop_words]
    # Join the filtered tokens back into a string
    text = ' '.join(filtered_tokens)
    
    return text
```

## Lemmatization 
### To return the word to it's basic form

```python
def lemmatize_tokens(tokens):
    # initialize WordNetLemmatizer
    lemmatizer = WordNetLemmatizer()

    # Lemmatize each word in the list of tokens
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]

    # make the words into sentence instead of a list
    return ' '.join(lemmatized_tokens)
```

```python
# apply function on tokenized column
df['lemmatization_column']=df['tokenized_column'].apply(lemmatize_tokens)
```

```python
# make new dataframe containing only needed columns for bert model
new_df = df.loc[:, ['lemmatization_column','offensive']]

new_df
```

### keras 

```python
# Split the data into input text and labels of the column offensive and get value as numpy array
text = new_df['lemmatization_column'].values
labels = new_df['offensive'].values

# Convert labels to binary matrix
labels = to_categorical(labels)
```

```python
# Split the data train and test 
train_text, test_text, train_labels, test_labels = train_test_split(text, labels, test_size=0.2, random_state=42)
```

```python
# download pretrained model for BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = TFBertModel.from_pretrained('bert-base-uncased')
```

```python
# Maximum sequence length for BERT
max_length = 128  

# tokenize the text train and test data
train_encodings = tokenizer.batch_encode_plus(train_text.tolist(), max_length=max_length, padding=True, truncation=True)
test_encodings = tokenizer.batch_encode_plus(test_text.tolist(), max_length=max_length, padding=True, truncation=True)

# np.array(encodings['input_ids']) NumPy array contains the tokenized input sequences converted into integer
# np.array(train_encodings['attention_mask']) NumPy array contains the attention mask 
train_inputs = [np.array(train_encodings['input_ids']), np.array(train_encodings['attention_mask'])]
test_inputs = [np.array(test_encodings['input_ids']), np.array(test_encodings['attention_mask'])]
```

```python
# input_ids and attention_mask are the input layers in the neural networks and input length is 128
input_ids = Input(shape=(max_length,), dtype=tf.int32, name="input_ids")
attention_mask = Input(shape=(max_length,), dtype=tf.int32, name="attention_mask")

# put the input layers inn the pretrained model
outputs = bert_model(input_ids, attention_mask=attention_mask)

# extracts the pooled output from the BERT model output
pooled_output = outputs[1]

# output layer is 2 nodes with softmax as an activation function
output = Dense(2, activation='softmax')(pooled_output)  

# define model input and output
model = Model(inputs=[input_ids, attention_mask], outputs=output)
```

```python
# compile using ADAM optimizer and loss function is cross entropy
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# fit 1 epoch and batch size equal 16
model.fit(train_inputs, train_labels, epochs=1, batch_size=16)  
```

```python
# predict using test data
predictions = model.predict(test_inputs)

# get the higger probabilty to assign the sentence to it from predicted data
predicted_labels = np.argmax(predictions, axis=1)

# get the higger probabilty to assign the sentence to it from actual data
true_labels = np.argmax(test_labels, axis=1)
```

```python
from sklearn.metrics import accuracy_score

# Calculate accuracy
accuracy = accuracy_score(true_labels, predicted_labels)

print(f"Accuracy: {accuracy}")
```
